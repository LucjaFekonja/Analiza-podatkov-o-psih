## 1. Sta višina in teža povezani z življensko dobo psa?
Poberimo podatke o višini, teži in življenski dobi.

### Starost in višina

Definirajmo povprečno vrednost in naredimo tabelo povprečnih vrednosti življenskih dob, višin in tež.

```python
def povprecna_vrednost (a, b):         # Zakaj mean() ne zazna kot funkcije (tudi, če mam statistics importan)
    return (a + b) / 2

psi.loc[:, 'pov_starost'] = povprecna_vrednost (psi['min_starost'], psi['max_starost'])
tabela_starosti = psi[['pasma', 'pov_starost']]

psi.loc[:, 'pov_visina'] = povprecna_vrednost(psi['min_visina'], psi['max_visina'])
tabela_visin = psi[['pasma', 'pov_visina']]

visina_starost = pd.merge(tabela_starosti, tabela_visin, on='pasma')
visina_starost
```

Izločimo vse vrstice, ki imajo katero izmed vrednosti Nan.

```python
visina_starost = visina_starost.dropna()
visina_starost
```

Narišimo graf spreminjanja življenske dobe glede na višino.

```python
visina_starost.sort_values('pov_visina').plot.scatter(x='pov_visina', y='pov_starost')
```

Iz grafa je razvidno, da višina nima bistvenega vpliva na življensko dobo psa, saj večina pasem živi med 12 in 14 let ne glede na višino. Nekoliko krajšo življensko dobo pa imajo največje pasme.

### Starost in teža

Naredimo tabelo povprečnih starosti in tež pasem.

```python
psi.loc[:, 'pov_teza'] = povprecna_vrednost(psi.min_teza, psi.max_teza)
tabela_tez = psi[['pasma', 'pov_teza']]

teza_starost = pd.merge(tabela_tez, tabela_starosti, on='pasma')
teza_starost
```

Izločimo vse vrednosti, ki imajo katero izmed vrednosti Nan.

```python
teza_starost.dropna()
```

Narišimo graf povprečne življenske dobe v odvisnosti od teže.

```python
teza_starost.sort_values('pov_teza').plot.scatter(x='pov_teza', y='pov_starost')
```

Vidimo, da lažje pasme dosežejo večje starosti.

## 4. Ali lahko iz opisa razberemo skupino v katero pes spada?

#### Verjetnost, da je pes iz skupine.

```python
st_psov_v_skupini = psi.pivot_table(index=['skupina'], aggfunc='size')
st_psov_v_skupini
```

```python
vsi_psi = len(psi)
vsi_psi
```

```python
verjetnost_skupine = st_psov_v_skupini / vsi_psi
verjetnost_skupine
```

#### Verjetnost, da je beseda značilna za skupino

Tabela skupin in opisov.

```python
skupina_opis = psi[['skupina', 'opis']].dropna()
skupina_opis 
```

Funkcija, ki poišče koren besed.

```python
def beseda_brez_koncnice(beseda):
    nova_beseda = ''.join(znak for znak in beseda if znak.isalpha())
    if not beseda:
        return ''
    else:
        return nova_beseda.rstrip('ds').rstrip('aeiou')
```

```python
import re
def spremenjen_niz(niz):
    seznam = [beseda_brez_koncnice(beseda) for beseda in niz.lower().split(' ')]
    return re.sub('\s+', ' ', ' '.join(seznam))
```

Preverim, da funkcija dela na eni izmed vrstic.

```python
spremenjen_niz("That of a smart, upstanding, strongly built moderate gundog bred for all types of shooting, especially for water-fowling. Great intelligence is combined with rugged endurance and a bold, dashing eagerness of temperament. Distinguishing characteristics are a topknot of long, loose curls and a body covered with a dense, crisply curled liver colored coat contrasted by a smooth face and a smooth &#8220;rat&#8221; tail.")
```

Naredimo tabelo, ki vsebuje podatek o skupini in korene vseh besed iz opisa.

```python
seznam_korenov = skupina_opis.opis.apply(spremenjen_niz).str.split(expand=True)
seznam_korenov
```

```python
tabela_skupina_koren = pd.merge(skupina_opis['skupina'], seznam_korenov, left_index=True, right_index=True)
tabela_skupina_koren
```

Iz te tabele naredimo tabelo s stolpcema 'skupina' in 'koren'.

```python
tabela = tabela_skupina_koren.set_index('skupina').stack().reset_index()[['skupina', 0]].rename(columns={'skupina': 'skupina', 0: 'koren'})
tabela
```

Tabela, ki prešteje število pojavitev korenov v opisih.

```python
st_korenov_v_skupini = pd.crosstab(tabela.koren, tabela.skupina)
st_korenov_v_skupini
```

Tabela verjetnosti, da je beseda v opisu.

```python
verjetnost_korena_ce_skupina = (st_korenov_v_skupini + 0.01) / st_psov_v_skupini
verjetnost_korena_ce_skupina
```

#### Ali lahko iz opisa labradorca sklepamo o skupini v katro spada?

Naredimo Series, ki vsebuje korene besed iz opisa labradorca.

```python
lab = pd.Series(spremenjen_niz("The Labrador Retriever is a strongly built, medium-sized, short-coupled, dog possessing a sound, athletic, well-balanced conformation that enables it to function as a retrieving gun dog; the substance and soundness to hunt waterfowl or upland game for long hours under difficult conditions; the character and quality to win in the show ring; and the temperament to be a family companion. Physical features and mental characteristics should denote a dog bred to perform as an efficient Retriever of game with a stable temperament suitable for a variety of pursuits beyond the hunting environment.").split(' '))
lab
```

Preverimo katere izmed besed iz tabele vseh korenov so v opisu labradorca.

```python
verjetnost_korena_ce_skupina.index.isin(lab)
```

Vzemimo tiste vrednosti iz tabele vrednosti vseh korenov, ki vsebujejo besede iz opisa labradorca.

```python
verjetnost_korena_ce_skupina[verjetnost_korena_ce_skupina.index.isin(lab)]
```

Z uporabo Bayesovega izreka izračunajmo verjetnost, da je labradorec v določeni skupini.

```python
(verjetnost_korena_ce_skupina[verjetnost_korena_ce_skupina.index.isin(lab)].prod() * verjetnost_skupine).sort_values()
```
